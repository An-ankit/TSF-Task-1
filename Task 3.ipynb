# EXPLORATORY DATA ANALYSIS

```python
# visualizing the dataset as a whole using the pair plot
import seaborn as sns 
sns.pairplot(data)
```

```python
# finding the pairwise correlations between the columns and visualising using heatmaps
data.corr()
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

Visualising the categories 

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(data['Category'].value_counts(), labels=data['Category'].value_counts().index,autopct='%0.2f%%',textprops = textprops)
plt.show()
```

```python
plt.figure(figsize= (10,16))
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.legend()
plt.show()
```

```python
# computing top categories in terms of sales from first 100 observations
top_category_sales = data.groupby("Category").Sales.sum().nlargest(n=100)
# computing top categories in terms of profit from first 100 observations
top_category_profit = data.groupby("Category").Profit.sum().nlargest(n=100)

# plotting to see it visually
plt.style.use('seaborn')
top_category_sales.plot(kind = 'bar',figsize = (10,5),fontsize = 15, color='green')
top_category_profit.plot(kind = 'bar',figsize = (10,5),fontsize = 15,color='red')
plt.xlabel('Category',fontsize = 20)
plt.ylabel('Total Sales/Profits',fontsize = 20)
plt.title("Top Category Sales vs Profit",fontsize = 35)
plt.legend()
plt.show()
```

Visualising the sub categories 

```python
# computing top sub-categories in terms of sales from first 100 observations
top_subcategory_sales = data.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
# computing top sub-categories in terms of profit from first 100 observations
top_subcategory_profit = data.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_sales.plot(kind = 'bar',figsize = (10,5),fontsize = 15, color='green')
top_subcategory_profit.plot(kind = 'bar',figsize = (10,5),fontsize = 15, color = 'red')
plt.xlabel('Sub-Category', fontsize = 20)
plt.ylabel('Total Sales/Profits',fontsize = 20)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 30)
plt.legend()
plt.show()
```

```python
# A more detailed view
plt.figure(figsize=(15,12))
plt.ylabel('Total Sales/Profits',fontsize = 20)
plt.xlabel('Sub-Category', fontsize = 20)
statewise = data.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.bar()
```

The above graph clearly shows that Copiers and Phones have the highest sales and profit and tables have highest loss.

Visualizing the Sales vs Profits in different Regions

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Region')
plt.pie(data['Region'].value_counts(), labels=data['Region'].value_counts().index,autopct='%1.1f%%',textprops=textprops)
plt.show()
```

The graph shows that West and East have same Profit though sales in the East are less as compared to West.

Visualising the Sales vs Profits in different states

```python
# computing top states in terms of sales from first 10 observations
top_states_sales = data.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
top_states_profit = data.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_sales.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color= 'green')
top_states_profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 30)
plt.legend()
plt.show()
```
